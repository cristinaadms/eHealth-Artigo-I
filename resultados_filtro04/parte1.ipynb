### PARTE 1: Arquitetura e Abordagem Computacional

#### 1.  Imports e carregamento dos dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from matplotlib.colors import LinearSegmentedColormap
import os

data_path = os.path.join('..', 'data', 'filtro04_processado.csv')

df = pd.read_csv(data_path)

print(f'Dataset carregado: {df.shape[0]} linhas x {df.shape[1]} colunas')
df.head(2)
```

#### 2.  Padronização global de estilo

```python
PALETTE = [
    '#3A6B8A',
    '#C8773A',
    '#4A7C59',
    '#5C4E8A',
    '#7A9E9F',
]

HEATMAP_CMAP = LinearSegmentedColormap.from_list(
    'custom_heatmap',
    [ "#6CB3DC", '#5C4E8A'],
    N=256
)

FIG_SIZE = (12, 6)

plt.rcParams.update({
    'font.family'                : 'serif',
    'font.serif'                 : ['Latin Modern Roman', 'DejaVu Serif'],
    'mathtext.fontset'           : 'cm',
    'axes.formatter.use_mathtext': True,

    'font.size'       : 13,
    'axes.titlesize'  : 16,
    'axes.labelsize'  : 14,
    'xtick.labelsize' : 12,
    'ytick.labelsize' : 12,

    'axes.linewidth'    : 1.2,
    'xtick.major.width' : 1.1,
    'ytick.major.width' : 1.1,
    'xtick.major.size'  : 4,
    'ytick.major.size'  : 4,

    'axes.grid'        : False,
    'figure.dpi'       : 150,
    'savefig.dpi'      : 300,
    'savefig.bbox'     : 'tight',
    'figure.facecolor' : 'white',
    'axes.facecolor'   : 'white',
})

print('✔  Estilo padronizado aplicado com sucesso.')
```

#### 3. Função auxiliar para gráfico de barras

```python
def plot_bar(counts, title, xlabel, ylabel='',
             color_cycle=None, filename=None):
    """
    Parâmetros:
    -------------------------------------------------------------------------
    counts      : pd.Series  — index = categorias, values = contagens
    title       : str        — título do gráfico
    xlabel      : str        — rótulo do eixo X
    ylabel      : str        — rótulo do eixo Y (padrão 'Frequência')
    color_cycle : list       — lista de cores HEX (padrão: PALETTE global)
    filename    : str        — se informado, salva a figura em PNG
    """
    if color_cycle is None:
        color_cycle = PALETTE

    n      = len(counts)
    colors = [color_cycle[i % len(color_cycle)] for i in range(n)]

    fig, ax = plt.subplots(figsize=FIG_SIZE)

    bars = ax.bar(counts.index, counts.values,
                  color=colors, width=0.65,
                  edgecolor='white', linewidth=0.6)

    # rótulos sobre as barras
    for bar in bars:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2,
                h + 0.12,
                str(int(h)),
                ha='center', va='bottom',
                fontsize=12)

    ax.set_title(title, pad=14)
    ax.set_xlabel(xlabel, labelpad=8)
    ax.set_ylabel(ylabel, labelpad=8)
    ax.yaxis.set_major_locator(mticker.MaxNLocator(integer=True))

    rotation = 35 if n > 4 else 0
    ha       = 'right' if n > 4 else 'center'
    ax.set_xticks(range(n))
    ax.set_xticklabels(counts.index, rotation=rotation, ha=ha)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.tight_layout()
    plt.show()
    plt.close(fig)

print('✔  Função plot_bar() definida.')
```

#### 4. Figura 1: Tipos de Arquitetura de Controle

```python
arch_cols = {
    'Centralizado'   : 'Centralizado',
    'Descentralizado': 'Descentralizado',
    'Distribuído '   : 'Distribuído',    
    'Hierárquico '   : 'Hierárquico',    
    'Multidomínio'   : 'Multidomínio',
}

arch_counts = (
    df[list(arch_cols.keys())]
    .sum()
    .rename(index=arch_cols)
    .sort_values(ascending=False)
    .astype(int)
)

plot_bar(arch_counts,
         title='Tipos de Arquitetura',
         xlabel='')
```

#### 5. Figura 2: Paradigmas Computacionais

```python
paradigms = (
    df['Paradigma Computacional']
    .dropna()
    .str.split(';')
    .explode()
    .str.strip()
    .value_counts()
)

plot_bar(paradigms,
         title='Paradigmas Computacionais',
         xlabel='')
```

#### 6. Figura 3: Técnicas Computacionais

```python
techniques = (
    df['Técnica Computacional']
    .dropna()
    .str.split(';')
    .explode()
    .str.strip()
    .value_counts()
)

plot_bar(techniques,
         title='Técnicas Computacionais',
         xlabel='')
```

#### 7. Figura 4: Mapa de Calor — Arquitetura vs Paradigmas

```python
df_exp = df.copy()
df_exp['Paradigma Computacional'] = (
    df_exp['Paradigma Computacional'].str.split(';')
)
df_exp = df_exp.explode('Paradigma Computacional')
df_exp['Paradigma Computacional'] = (
    df_exp['Paradigma Computacional'].str.strip()
)

arch_rename = {
    'Centralizado'   : 'Centralizado',
    'Descentralizado': 'Descentralizado',
    'Distribuído '   : 'Distribuído',
    'Hierárquico '   : 'Hierárquico',
    'Multidomínio'   : 'Multidomínio',
}
df_exp      = df_exp.rename(columns=arch_rename)
arch_labels = list(arch_rename.values())

df_melt = df_exp.melt(
    id_vars    = 'Paradigma Computacional',
    value_vars = arch_labels,
    var_name   = 'Arquitetura',
    value_name = 'Presente'
)
df_melt = df_melt[df_melt['Presente'] == 1]

heat_df = (
    df_melt
    .groupby(['Paradigma Computacional', 'Arquitetura'])
    .size()
    .unstack(fill_value=0)
    .reindex(columns=arch_labels, fill_value=0)
)

# 5. Plot
fig, ax = plt.subplots(figsize=FIG_SIZE)

im = ax.imshow(heat_df.values, aspect='auto',
               cmap=HEATMAP_CMAP, vmin=0)

cbar = fig.colorbar(im, ax=ax, shrink=0.85, pad=0.02)
cbar.set_label('', fontsize=13)
cbar.ax.tick_params(labelsize=11)

ax.set_xticks(range(len(arch_labels)))
ax.set_xticklabels(arch_labels, rotation=0, ha='center', fontsize=12)

ax.set_yticks(range(len(heat_df.index)))
ax.set_yticklabels(heat_df.index, fontsize=11)

ax.set_xlabel('',        labelpad=8, fontsize=14)
ax.set_ylabel('',         labelpad=8, fontsize=14)
ax.set_title('Tipo de Arquitetura vs Paradigma Computacional',
             pad=14, fontsize=16)

for i in range(heat_df.shape[0]):
    for j in range(heat_df.shape[1]):
        val      = heat_df.values[i, j]
        norm_val = val / (heat_df.values.max() or 1)
        txt_color = 'white' if norm_val > 0.55 else '#2B2B2B'
        ax.text(j, i, str(val),
                ha='center', va='center',
                fontsize=12, color=txt_color)

ax.spines[:].set_visible(False)
plt.tight_layout()
plt.show()
plt.close(fig)
```
